```python
import google.generativeai as genai
import PIL.Image as pilimg
from google.colab import userdata
from google.generativeai.generative_models import GenerativeModel
from google.generativeai.types.generation_types import GenerateContentResponse
```

```python
GEMINI_API_KEY: str = userdata.get('GEMINI_API_KEY')
```

```python
if GEMINI_API_KEY:
  print("API key fetched successfully!")
else:
  print("API key not found. Please set the 'OPENAI_API_KEY' user data.")
```

```python
genai.configure(api_key=GEMINI_API_KEY)
```

- Global Configuration of Model for this specific POC

```python
# Model Selection/Configuration

model: GenerativeModel = genai.GenerativeModel(
    model_name = 'gemini-2.0-flash-exp',
    generation_config = genai.GenerationConfig(
        max_output_tokens = 2500,
        temperature = 0.7,
    ))

print(f"Successfully got access to: {model.model_name}")
```

```python
image = "/content/drive/MyDrive/Photo from Alpha.ai"

img = pilimg.open(image)
display(img)    # display is undefined, but working!
```

```python
prompt = """What do you see in this image? Explain in bullet points, also create a table to later store in the relational database"""
```

- Pass in the prompt and image to get the prompt completed.

```python
response: GenerateContentResponse = model.generate_content([prompt, img])
print(response.text)
```
